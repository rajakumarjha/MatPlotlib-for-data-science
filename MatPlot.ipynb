```python
import matplotlib.pyplot as plt
#printing a list
```

```python
plt.plot([1,2,3,4,0])
plt.ylabel("Number")
plt.xlabel("Indices")
plt.title("First Plot")
plt.show()
```

```python
plt.plot([1,5,3,4],[2,5,8,6])
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title("X vs Y")
#can be made like a grid
plt.grid()
plt.show()
```

```python
plt.plot([1,5,3,4],[2,5,8,6],'ro')#can be colored accordingly
```

```python
import numpy as np
a=np.arange(0,2,.2)
plt.plot(a,a**2,'rs',label='^2')
plt.plot(a,a**2.5,'bo',label='^2.5')
plt.plot(a,a**5,'b',label='^5')
plt.grid()
plt.legend() # They appear on the top,it shows the scales 
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title("X vs Y")
plt.show()
```

```python
a=np.arange(-10,10,.1)
s=np.sin(a)
plt.plot(s)
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title("Sin")
plt.show()
```

```python
def f(t):
    return np.exp(t)
t=np.arange(-10,10,.10)
t1=np.arange(0,5,.12)
plt.subplot(2,1,1)
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title("function")
plt.plot(f(t),t,'r')

plt.subplot(2,1,2)
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title("Sin")
plt.plot(t,np.sin(t),'g')

plt.show()
```

### This matplot has many use in data science specially when we want to plot some function or some data to see some kind of relationships there.
